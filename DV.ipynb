```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Sample Data
data = {
'Caterogy' : ['A', 'B', 'C', 'D', 'E'],
'Values 1' : [10, 20, 30, 25, 15],
'Values 2' : [5, 15, 25, 20, 10],
}
df = pd.DataFrame(data)
```

```python
#Line chart
plt.figure(figsize =(6,4))
plt.plot(df['Caterogy'], df['Values 1'], marker ='o', linestyle ='-', label ='Line 1')
plt.plot(df['Caterogy'], df['Values 2'], marker ='s', linestyle ='--', label ='Line 2')
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Line Chart")
plt.legend()
plt.grid()
plt.show()
```

```python
#BAR CHART
plt.figure(figsize =(6,4))
plt.bar(df['Caterogy'], df['Values 1'], color = 'skyblue', label ='Values 1')
plt.bar(df['Caterogy'], df['Values 2'], color = 'orange', bottom = df['Values 1'], label ='Values 2')
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Stacked Bar Plot")
plt.legend()
plt.show()
```

```python
#SCATTER PLOT
plt.figure(figsize =(6,4))
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
plt.scatter(x, y, c='red', alpha = 0.6, edgecolors='black')
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("Scatter Plot")
plt.show()
```

```python
#BOX PLOT
plt.figure(figsize =(6,4))
sns.boxplot(data = [df['Values 1'], df['Values 2']], palette='Set2')
plt.xticks([0,1], ['Values 1', 'Values 2']) #Customize
plt.title("Box Plot")
plt.show()
```

```python
#HEAT MAP
heat_data = np.random.rand(5,5) #Generated random data (5X5 matrix)
plt.figure(figsize = (6,4))
sns.heatmap(heat_data, annot = True, cmap = 'coolwarm', linewidth =0.5)
plt.title("Heatmap")
plt.show()
```

```python
#HISTOGRAM
plt.figure(figsize = (6,4))
data = np.random.randn(1000)
plt.hist(data, bins = 30, color = 'purple', edgecolor = 'black', alpha =0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
#Creating multiple subplots
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load built-in dataset from seaborn
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

#Create a figure with multiple subplots
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12,10))

#Scattter plot - Tips Dataset
sns.scatterplot(ax = axes[0,0], data = tips, x = "total_bill", y = "tip", hue = "sex", palette = "viridis")
axes[0,0].set_title("Total Bill vs Tip")

#Boxplot = Iris Dataset
sns.boxplot(ax = axes[0,1], data = iris, x = "species", y = "sepal_length", hue ="petal_width", palette=
"pastel")
axes[0,1].set_title("Sepal Length Distribution by Species")

#Histrogram - Tips Dataset
sns.histplot(ax=axes[1,0], data = tips, x = "total_bill", bins = 20, color = "purple", kde = True)
axes[1,0].set_title("Distribution of Total Bills")

#Heatmap - Correlation Matrix of Iris Dataset
corr = iris.select_dtypes(include = ['number']).corr() #includes onlu numerical cols
sns.heatmap(ax = axes[1,1], data = corr, annot = True, cmap = "coolwarm", linewidths = 0.5)
axes[1,1].set_title("Iris Dataset Correlation Matrix")

plt.tight_layout()
plt.show()
```

```python
#IRIS DATASET
#Create a figure with multiple subplots
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12,10))

#Scattter plot - Tips Dataset
sns.scatterplot(ax = axes[0,0], data = iris, x = "species", y = "sepal_length", hue = "petal_width",
palette = "viridis")
axes[0,0].set_title("Sepal Length Distribution by Species")

#Boxplot = Iris Dataset
sns.boxplot(ax = axes[0,1], data = iris, x = "species", y = "sepal_length", hue ="petal_width", palette=
"pastel")
axes[0,1].set_title("Sepal Length Distribution by Species")

#Histrogram - Iris Dataset
sns.histplot(ax=axes[1,0], data = iris,  x = "species", bins = 20, color = "purple", kde = True)
axes[1,0].set_title("Sepal Length Distribution by Species")

#Heatmap - Correlation Matrix of Iris Dataset
corr = iris.select_dtypes(include = ['number']).corr() #includes onlu numerical cols
sns.heatmap(ax = axes[1,1], data = corr, annot = True, cmap = "coolwarm", linewidths = 0.5)
axes[1,1].set_title("Iris Dataset Correlation Matrix")

plt.tight_layout()
plt.show()
```

```python
#PIE CHART
# Pie chart - Tips Dataset
plt.figure(figsize = (6,6))
tips_day_counts = tips["day"].value_counts()  #working on day column
plt.pie(tips_day_counts, labels = tips_day_counts.index, autopct = '%1.1f%%', startangle = 90,
colors=["lightgreen", "lightblue", "orange", "pink"])

#Automatic Percentage --> displays the percentage values inside the slices of a pie chart
plt.title("Pie Chart of Tips by Day")
plt.show()

#Muliple Subplots - Pie, Pair, and Violin
fig , axes = plt.subplots(nrows = 1, ncols = 3, figsize = (18,6))

#Pie Chart
axes[0].pie(tips_day_counts, labels = tips_day_counts.index, autopct = '%1.1f%%',
colors=["lightgreen", "lightblue", "orange", "pink"])
axes[0].set_title("Pie Chart of Tips by Day")

sns.pairplot(iris, hue = "species", palette="muted")
plt.suptitle("Pair Plot of Iris Dataset", fontsize = 16, y = 1.02)

sns.violinplot(ax = axes[2], data = tips, x = "day", y= "total_bill", hue = "sex", split = True ,palette =
"muted")
axes[1].set_title("Pair Plot of Tips Dataset")

plt.tight_layout()
plt.show()
```
